```python
#@title Código 3
r = requests.get(url)
# print json content
#print(r.json())
df = pd.read_csv('https://raw.githubusercontent.com/francisconetodata/loteria_jogo/main/MV_teste.csv')
```

# Resultado

O concurso da mega da virada 2021/22 é 2440 - Verificar se o número do concurso coincide com o mostrado abaixo:

```python
#@title Dados do concurso 
print('Resultado da Mega Sena')
print('\nConcurso:')
print(r.json()['numero_concurso'])
print('\nResultado:')
print(r.json()['dezenas'])
```

# Tratamento

```python
#@title Tratamento de dados { display-mode: "form" }
resultados = (r.json()['dezenas'])
int_list = list(map(int, resultados))
resultado = int_list
#resultado
numeros = df.drop(columns='Jogo')
vetor_resultado = []
vetor_nome_resultado = []
colunas = numeros.columns
for i in range(len(numeros)):
  contador = 0
  for j in colunas:
    if numeros[f'{j}'][i] in resultado:
      contador = contador + 1
  vetor_resultado.append(contador)
  if contador == 0:
    vetor_nome_resultado.append('Nenhum Acerto')
  elif contador == 1:
    vetor_nome_resultado.append('Um Acerto')
  elif contador == 2:
    vetor_nome_resultado.append('Dois Acertos')
  elif contador == 3:
    vetor_nome_resultado.append('Terno')
  elif contador == 4:
    vetor_nome_resultado.append('Quadra')
  elif contador == 5:
    vetor_nome_resultado.append('Quina')
  else:
    vetor_nome_resultado.append('Mega!!!')


numeros['Quantidade acertos'] = vetor_resultado
numeros['Nome do resultado'] = vetor_nome_resultado
topresulta = numeros.groupby('Nome do resultado').agg({ 'Quantidade acertos':'count'}).sort_values(by='Quantidade acertos' )
```
